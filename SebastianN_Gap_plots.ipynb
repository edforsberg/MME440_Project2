```python
from sklearn.cluster import KMeans, SpectralClustering
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import CreateDataFunctions as df
import gap

import warnings
warnings.filterwarnings("ignore")
```

```python
data1, c1 = df.Data_gussian_cluster(nr_data_points = 1000)
data2, c2 = df.Data_separated_clusters()
data3, c3 = df.Gaussian_cluster_grid(grid_size=10, scale=8, nr_data_points=10, sigma=0.6)
data4, c4 = df.ThreeCirclesData()
data5_grid = 2
data5_nr_features = 5
data5, c5 = df.Gaussian_cluster_grid(grid_size=data5_grid, scale=25, nr_data_points=10, nr_features=data5_nr_features, sigma=0.5)
data6, c6 = df.Data_gussian_cluster_hard(nr_data_points = 500)
```

```python
wk_1, wk_ref1, sk1, Ks1 = gap.get_gap_statistic(data1, verbose=True)
wk_2, wk_ref2, sk2, Ks2 = gap.get_gap_statistic(data2, verbose=True)
wk_3, wk_ref3, sk3, Ks3 = gap.get_gap_statistic(data3, K_max=110, verbose=True)
wk_4, wk_ref4, sk4, Ks4 = gap.get_gap_statistic(data4, verbose=True)
wk_5, wk_ref5, sk5, Ks5 = gap.get_gap_statistic(data5, K_max=32, verbose=True)
wk_6, wk_ref6, sk6, Ks6 = gap.get_gap_statistic(data6, verbose=True)
```

```python
def plot_wk(Ks, wk, wk_ref, optk):
    plt.axvline(x=optk, color='r', linestyle='dashed', label='Optimal K')
    plt.plot(Ks, wk, label=r"$log W_k$")
    plt.plot(Ks, wk_ref, label=r"$\frac{1}{B}\sum^{B}_{b}\log W_{kb}$")
    plt.legend(loc='upper right')
    plt.xlabel("K")


    #ax2.plot(Ks, np.exp(wk))
    #ax2.set_xlabel("K")
    #ax2.set_ylabel(r"$W_k$")
    

    plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = (8.0, 6.0)
optk1 = gap.get_optimal_k(wk_1, wk_ref1, sk1, Ks1)
plot_wk(Ks1, wk_1, wk_ref1, optk1)
print("Optimal K {}".format(optk1))

plt.show()

optk2 = gap.get_optimal_k(wk_2, wk_ref2, sk2, Ks2)
plot_wk(Ks2, wk_2, wk_ref2, optk2)
print("Optimal K {}".format(optk2))

optk3 = gap.get_optimal_k(wk_3, wk_ref3, sk3, Ks3)
plot_wk(Ks3, wk_3, wk_ref3, optk3)
print("Optimal K {}".format(optk3))

optk4 = gap.get_optimal_k(wk_4, wk_ref4, sk4, Ks4)
plot_wk(Ks4, wk_4, wk_ref4, optk4)
print("Optimal K {}".format(optk4))

optk6 = gap.get_optimal_k(wk_6, wk_ref6, sk6, Ks6)
plot_wk(Ks6, wk_6, wk_ref6, optk6)
print("Optimal K {}".format(optk6))
```
